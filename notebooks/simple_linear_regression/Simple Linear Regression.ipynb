```python
import pandas
import numpy
from matplotlib import pyplot
```

```python
dataset = pandas.read_csv("Salary_Data.csv")
```

```python
x = dataset.iloc[:, :-1]
y = dataset.iloc[:, 1]
```

```python
pyplot.scatter(x,y)
```

```python
from sklearn.cross_validation import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/3, random_state = 0)
```

```python
y_train
```

```python
pyplot.scatter(x_train, y_train)
```

```python
from sklearn.linear_model import LinearRegression
r = LinearRegression()
r.fit(x_train, y_train)
```

```python
y_pred = r.predict(x_test)
```

```python
pyplot.scatter(range(len(y_test)), y_test)
pyplot.scatter(range(len(y_pred)), y_pred)
```

```python
pyplot.scatter(x_train, y_train, color='red')
pyplot.scatter(x_test, )
pyplot.plot(x_train, r.predict(x_train))
```
